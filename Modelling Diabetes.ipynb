```python
import pandas as pd
```

```python
data=pd.read_csv("diabetes.csv")
```

```python
data.head(10)
```

```python
df=data.copy()
```

```python
df=df.drop(["Pregnancies","DiabetesPedigreeFunction"], axis=1)
```

```python
df.head(10)
```

```python
#Using Variable Baru


# Separating X and y
X = df.drop(['Outcome'], axis=1)
Y = df['Outcome']

# Build random forest model
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X, Y)

# Saving the model
import pickle
pickle.dump(clf, open('diabetes_clf.pkl', 'wb'))
```
